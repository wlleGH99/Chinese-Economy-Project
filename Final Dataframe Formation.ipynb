```python
rewritten_df = pd.DataFrame()
old_cols = to_modify.columns.values
unknown = [to_modify.columns.values[9], to_modify.columns.values[10]]

index = 0

for col in new_vals:
    if col in old_cols:
        rewritten_df[col] = to_modify[col].values
    else:
        if index == 0:
            rewritten_df['Pub_MD_Type of Coverage - Article Level (lva)'] = to_modify[unknown[index]].values
        elif index == 1:
            rewritten_df['Pub_MD_Type of Coverage - Source Level (lvs)'] = to_modify[unknown[index]].values
        index+=1
        
rewritten_df
```

```python
import numpy as np

ret = []
df_code = []
df_list = []

for codes in rewritten_df['Article_MD_subject_codes']:
    code_list = []
    coding = 100
    if isinstance(codes, str):
        codes = codes.lower()
        if 'npag' in codes:
            coding = 0
            code_list.append(0)
        if 'nedi' in codes:
            coding = 1
            code_list.append(1)
        if 'nedc' in codes:
            coding = 2
            code_list.append(2)
        if 'nlet' in codes:
            coding = 3
            code_list.append(3)
        if 'nrvw' in codes:
            coding = 4
            code_list.append(4)
        if 'nobt' in codes or 'nadvtr' in codes:
            coding = 5
            code_list.append(5)
    if coding != 100:
        df_code.append(min(code_list))
        df_list.append(code_list)
    else:
        df_code.append(np.nan)
        df_list.append(np.nan)
    
print('Done')
```

```python
rewritten_df['Article_Type'] = df_code
rewritten_df['Article_Type_List'] = df_list
```
